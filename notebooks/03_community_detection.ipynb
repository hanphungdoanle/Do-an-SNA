```python
# === CELL 1: IMPORT THƯ VIỆN & KẾT NỐI GOOGLE DRIVE ===
import networkx as nx
import pandas as pd
import pickle
import os
import community.community_louvain as community_louvain

# Kết nối với Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Cài đặt đường dẫn chính xác tới thư mục SNA
WORK_DIR = '/content/drive/MyDrive/Colab Notebooks/SNA/'
input_path = os.path.join(WORK_DIR, 'amazon_graph.gpickle')
output_path = os.path.join(WORK_DIR, 'cluster_labels.csv')

print("Đã import thành công các thư viện và thiết lập đường dẫn!")
```

```python
# === CELL 2: ĐỌC DỮ LIỆU ĐỒ THỊ TỪ THÀNH ===
print(f"\nĐang đọc dữ liệu từ file: {input_path}...")
try:
    with open(input_path, 'rb') as f:
        G_sub = pickle.load(f)
    print("Tải đồ thị thành công! ✅")
    print(f"Tổng số Node hiện tại: {G_sub.number_of_nodes():,}")
    print(f"Tổng số Edge hiện tại: {G_sub.number_of_edges():,}")
except FileNotFoundError:
    print("LỖI: Không tìm thấy file. Hãy kiểm tra lại xem file amazon_graph.gpickle đã thực sự nằm trong thư mục SNA trên Drive chưa nhé.")
```

```python
# === CELL 3: TRIỂN KHAI THUẬT TOÁN LOUVAIN ===
print("\nĐang chạy thuật toán Louvain để phân cụm cộng đồng (Community Detection)...")
# Hàm best_partition trả về một dictionary với key là Node ID và value là Cluster ID (Nhãn cụm)
partition = community_louvain.best_partition(G_sub)

# Lấy số lượng cụm (communities) được tạo ra
num_clusters = len(set(partition.values()))
print(f"Thuật toán đã chia mạng lưới thành {num_clusters} cụm sản phẩm khác nhau.")
```

```python
# === CELL 4: TÍNH CHỈ SỐ MODULARITY (Dành cho Chương 5 Báo Cáo) ===
print("\nĐang đánh giá chất lượng phân cụm bằng chỉ số Modularity...")
modularity_score = community_louvain.modularity(partition, G_sub)

print("="*50)
print("KẾT QUẢ ĐÁNH GIÁ (Ghi vào Báo Cáo):")
print(f"-> Chỉ số Modularity: {modularity_score:.4f}")
print("="*50)
# Lưu ý cho Báo cáo: Modularity > 0.3 thường cho thấy cấu trúc cộng đồng rõ ràng.
# Mức điểm này dùng để chứng minh thuật toán hoạt động hiệu quả.
```

```python
# === CELL 5: LƯU KẾT QUẢ VÀ BÀN GIAO CHO AN ===
print(f"\nĐang xuất kết quả ra file: {output_path}...")

# Chuyển đổi dictionary thành Pandas DataFrame để dễ dàng lưu thành file CSV
df_clusters = pd.DataFrame(list(partition.items()), columns=['Node_ID', 'Cluster_ID'])

# Có thể sort lại theo Cluster_ID cho đẹp
df_clusters = df_clusters.sort_values(by=['Cluster_ID', 'Node_ID']).reset_index(drop=True)

# Lưu thành CSV (không lấy cột index)
df_clusters.to_csv(output_path, index=False)

print("LƯU THÀNH CÔNG! ✅")
print("Hãy gửi file 'cluster_labels.csv' này cho An để bạn ấy vẽ biểu đồ nhé.")
```
