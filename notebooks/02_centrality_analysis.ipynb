```python
from google.colab import drive
import networkx as nx
import pandas as pd
import pickle

# Mount Google Drive
drive.mount('/content/drive')
GRAPH_PATH = '/content/drive/MyDrive/Colab Notebooks/Do-an-SNA/data/processed/amazon_graph.gpickle'
CENTRALITY_OUT = '/content/drive/MyDrive/Colab Notebooks/Do-an-SNA/data/processed/centrality_scores.csv'
TOP_PAGERANK_OUT = '/content/drive/MyDrive/Colab Notebooks/Do-an-SNA/data/processed/top_pagerank_nodes_for_report.csv'

# 1. Đọc file đồ thị
print("Đang tải đồ thị amazon_graph.gpickle...")
with open(GRAPH_PATH, 'rb') as f:
    G_sub = pickle.load(f)

print("Số nodes:", G_sub.number_of_nodes())
print("Số edges:", G_sub.number_of_edges())

# 2. Tính Centrality
print("Đang tính Degree Centrality...")
degree_cen = nx.degree_centrality(G_sub)

print("Đang tính Betweenness Centrality (vui lòng đợi vài phút)...")
betweenness_cen = nx.betweenness_centrality(G_sub)

print("Đang tính PageRank...")
pagerank_cen = nx.pagerank(G_sub)

# 3. Lưu kết quả
df_centrality = pd.DataFrame({
    'Node_ID': list(G_sub.nodes()),
    'Degree': [degree_cen[node] for node in G_sub.nodes()],
    'Betweenness': [betweenness_cen[node] for node in G_sub.nodes()],
    'PageRank': [pagerank_cen[node] for node in G_sub.nodes()],
    'Degree_Count': [G_sub.degree(node) for node in G_sub.nodes()]
})

top10_degree = df_centrality.nlargest(10, 'Degree')
top10_betweenness = df_centrality.nlargest(10, 'Betweenness')
top10_pagerank = df_centrality.nlargest(10, 'PageRank')

print("\n--- TOP 10 DEGREE ---")
print(top10_degree[['Node_ID', 'Degree', 'Degree_Count']])

print("\n--- TOP 10 BETWEENNESS ---")
print(top10_betweenness[['Node_ID', 'Betweenness']])

print("\n--- TOP 10 PAGERANK ---")
print(top10_pagerank[['Node_ID', 'PageRank', 'Degree_Count']])

df_centrality.to_csv(CENTRALITY_OUT, index=False)
top10_pagerank.to_csv(TOP_PAGERANK_OUT, index=False)

print("\nĐã xuất file:")
print(CENTRALITY_OUT)
print(TOP_PAGERANK_OUT)

```
