```python
import networkx as nx
import pandas as pd
import pickle

# 1. Đọc file đồ thị Thành gửi
print("Đang tải đồ thị amazon_graph.gpickle...")
with open('amazon_graph.gpickle', 'rb') as f:
    G_sub = pickle.load(f)

# your code here
print("Đang tính Degree Centrality...")
degree_cen = nx.degree_centrality(G_sub)

print("Đang tính Betweenness Centrality (Vui lòng đợi vài phút)...")
betweenness_cen = nx.betweenness_centrality(G_sub)

print("Đang tính PageRank...")
pagerank_cen = nx.pagerank(G_sub)

df_centrality = pd.DataFrame({
    'Node_ID': list(G_sub.nodes()),
    'Degree': [degree_cen[node] for node in G_sub.nodes()],
    'Betweenness': [betweenness_cen[node] for node in G_sub.nodes()],
    'PageRank': [pagerank_cen[node] for node in G_sub.nodes()]
})

top10_degree = df_centrality.nlargest(10, 'Degree')
top10_betweenness = df_centrality.nlargest(10, 'Betweenness')
top10_pagerank = df_centrality.nlargest(10, 'PageRank')

print("\n--- TOP 10 DEGREE ---")
print(top10_degree[['Node_ID', 'Degree']])

print("\n--- TOP 10 BETWEENNESS ---")
print(top10_betweenness[['Node_ID', 'Betweenness']])

print("\n--- TOP 10 PAGERANK ---")
print(top10_pagerank[['Node_ID', 'PageRank']])

df_centrality.to_csv('centrality_scores.csv', index=False)
print("\nĐã xuất file centrality_scores.csv thành công!")
# end your code here
```
